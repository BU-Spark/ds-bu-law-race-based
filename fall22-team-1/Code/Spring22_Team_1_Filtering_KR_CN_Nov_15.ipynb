```python
list_docs = list(range(2005,2023))
print(len(list_docs))
list_docs
```

```python
total_rows = 0

total_kr = 0
total_cn = 0
total_both = 0

head = None

for i in list_docs:
    
    file_name = str(i) + "_patent_grant_processed.csv"
    
    print("\n"+"-"*30 + file_name + "-"*30)

    with open("Team 1 Processed Data/"+file_name, 'r',encoding="utf8") as file:
        csvreader = csv.reader(file)
        
        count = 0
        
        china_count = 0
        korea_count = 0
        both_count = 0

        for row in csvreader:
            
            if count==0:
                print(len(row))
                if row!=head:
                    head = row
                    print("----------HEAD CHANGE----------")
            
            count+=1
            
            if count<5:
                print(row)
                print(row[9])
                print(row[1])
                
                
            if 'KR' in row[9] and 'CN' in row[9]:
                both_count+=1
                
            elif 'KR' in row[9]:
                korea_count+=1
            
            elif 'CN' in row[9]:
                china_count+=1            
          
    print("\n")
    print("No of rows in document-",count)
    print("No of both CN & KR in document-",both_count)
    print("No of CN in document-",china_count)
    print("No of KR in document-",korea_count)
    
    print("\n")
    
    total_rows += count
    total_both += both_count
    total_kr += korea_count
    total_cn += china_count

print("\n\n")
print("Total no of rows in all documents-",total_rows)
print("Total CN & KR in all documents-",total_both)
print("Total CN in all documents-",total_cn)
print("Total KR in all documents-",total_kr)

print(head)
```
